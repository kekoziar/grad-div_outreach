```python
# 1
import requests
from bs4 import BeautifulSoup
from datetime import datetime as dt
```

```python
# 2
url = "https://graduate.ucr.edu/need-help"
webpage = requests.get(url)
```

```python
# 3
soup = BeautifulSoup(webpage.content, 'html.parser')
#print(soup.prettify())
#print(soup.get_text())
#soup.find_all('div')
```

```python
# 4
outreach = []
for content in soup.find_all('div', attrs={'class': 'accordion-content'}):
    string = str(content)
#    print(string)
    fga = content.find('p') # fga <- faculty/graduate advisor
#    print(fga)
#    print("++++++++++++++++++++++++++")
    newstring = string[string.find(" id=""")+5:string.find(">")-1]+","+fga.text
    if(newstring.find("designated_emphasis")==-1):
        newstring = newstring.replace("\n","\n"+string[string.find(" id=""")+5:string.find(">")-1]+",")
        newstring = newstring.replace("Â ","")
        # newstring = newstring+"\n"
        print("++++++++++++++++++++++++++")
        print(newstring)
    #    print("++++++++++++++++++++++++++")
        outreach.append(newstring)
```

```python
# 5
temp = "\n".join(outreach)
```

```python
# 6
xml_file = open("./results/"+dt.now().strftime("%Y%m%d")+"_grad_contacts.csv", "w")
xml_file.write(temp)
xml_file.close()
# fin. go to file cleanup
```
